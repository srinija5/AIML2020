### interpolate():
 - Guessed the NaN value based on the other values in the column
 - Guessed value will be in the range of the values in the column
 - All the values filled won't be the same.
 - Only for numerical columns
 
### Difference Between interpolate() and fillna():
 - We're giving instructions to fill the specific value in fillna()
 - Whereas in interpolate() the function itself is choosing the value

### Handling missing values using 'Scikit-learn':
1. SimpleImputer

```python
#SimpleImputer:
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy='mean')
#SimpleImputer works only for numerical columns
```

```python
#It is a 2 step transformation
#imputer->object
imputer.fit(data.iloc[:,1:3]) #Indices of column 1 and 2
#Substituting the missing values of 1,2 columns with mean
```

```python
#To reflect the changes
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3])
data
```

4. Question 4:
Write a python program which finds the maximum number from num1 to num2 (num2 inclusive) based on the following rules.
Always num1 should be less than num2
Consider each number from num1 to num2 (num2 inclusive). Populate the number into a list, if the below conditions are satisfied
Sum of the digits of the number is a multiple of 3
Number has only two digits
Number is a multiple of 5
Display the maximum element from the list
In case of any invalid data or if the list is empty, display -1.





```python
#Q1:
def find_product(num1,num2,num3):
    product=0
    if (num1 != 7 and num2 != 7 and num3 != 7):
        product = num1 * num2 * num3
    elif (num1 == 7):
        product = num2 * num3
    elif(num2 == 7):
        product = num3
    elif(num3 == 7):
        product = -1
    return product

product=find_product(3,7,8)
print(product)
```
