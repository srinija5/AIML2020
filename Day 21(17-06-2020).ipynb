### Dummy variables using Scikit-Learn Library
 - Label Encoder:
     - Creating dummy variables: 
         - It is a class present in Library
         - Create Object of LabelEncoder
         - TO apply the object to the dataset(town), we use fit_transform
         - Syntax: objectname.fit_transform(datasetname.columnname)
 - OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() #Object name->le
```

```python
price_data=data.copy()
price_data
```

```python
price_data.town=le.fit_transform(data.town)
price_data
#String encoding is done. Transformed into 0,1,2->Dummy variable
```

### LabelEncoder Disadvantages:
 - As dummy variables are 0-.....1000Then the algorithm assumes that these are numericals then compares. Like- 9>3,3<1,etc
  - LavelEncoder is only usable when there are only 2 categories lik 0,1
  - Where 0->presence
  - 1->absence

### OneHotEncoder:
 - Similar to pd.get_dummies()
 - Steps:
     - Import OneHotEncoder class
     - Create an object for the class
     - ColumnTransformer ->Inside this we pass the OneHotEncoder
         - It has 2 mandatory arguments:
             - Transforms
             - Remainder
         - transormers argument-> Accepts a list of tuples

```python
#Importing libraries 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#NOw we pass OneHotEncoder
column_transformer=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],
                                     remainder='passthrough')
#Name of the transformer->encoder
#What is the transformer->OneHotEncoder
#which column->[0]->index
#Remainder->passthrough
```

```python
#applying to dataset
data =column_transformer.fit_transform(data)
data
```

```python
#converting the array into dataframe
X=pd.DataFrame(data)
X
```

### Feature scaling: (Not Mandatory for all algorithms)

```python
sample=pd.DataFrame({'W.E':[2,4,5,6,7,8],'Age':[28,25,32,45,33,32],
                    'Salary':[300000,400000,500000,600000,700000,800000],'Distance':[10,12,23,32,15,17],
                    'Expected Salary':[400000,500000,600000,700000,800000,900000]})
```

```python
sample
```

### Scale():

 - In most of the ML algorithms we use Distance based Formulas
 - Salary column is hugely dominating the other columns in terms of magnitude
 - Also all columns in dataset belong to different units.
 - Exp->years,Salary->Currency
 - Units and Scale(Magnitude)
 - Feature Scaling->When applied, this units and scaling will be removed
 - To make the data unitless and scaleless , we have to apply feature scaling

```python
from sklearn.preprocessing import scale
scaled_data=scale(sample)
scaled_data
```

```python
scaled_data=pd.DataFrame(scaled_data)
scaled_data
```

### StandardScaler():

```python
#Creating a copy of original data
standard_data=sample.copy()
standard_data
```

```python
#Using StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sample=sc.fit_transform(sample)
sample
#Whenever we apply the fit_transform the output will be in array form
```

```python
sample=pd.DataFrame(sample)
sample
```

### MinMaxScaler():

```python
#MinMaxScaler:
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(standard_data)
standard_data=ms.transform(standard_data)
standard_data
```
