### Image operations

```python
from scipy import misc
data = misc.face()
plt.imshow(data)
```

```python
data.shape,type(data)
```

```python
flip = np.flipud(data)
plt.imshow(flip)
```

```python
rand_att=np.random.randint(low=0,high=255,size=data.shape)
plt.imshow(rand_att)
```

```python
from scipy import ndimage
data_rotate=ndimage.rotate(data,135)
plt.imshow(data_rotate)
```

### Problem solving:

```python
#Q7:xyz : 750 , abc:265 , qwerty:195
s=input()
list1 = s.split(",")
list2 = [i.split(":")[1] for i in list1]
list3=[]
for num in list2 :#num = '750'
    n = int(num)#n=750
    rev = num[::-1]#rev=57
    while str(n)!= rev:
        n += int(rev)
        rev = str(n)[::-1]
    list3.append(n)
list1[list3.index(max(list3))].split(":")[0]
```
