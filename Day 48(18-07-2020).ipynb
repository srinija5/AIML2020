```python
#5 Clusters
from sklearn.cluster import AgglomerativeClustering
clust= AgglomerativeClustering(n_clusters=5,linkage='complete',affinity='euclidean').fit(norm_data)
```

```python
#Cluster labels
clust.labels_
```

## DataFrame for cluster labels:

```python
cluster=pd.DataFrame(clust.labels_)
cluster
```

## Creating a new column in original dataset:

```python
#create a new column 'cluster' in univ dataset
univ['cluster']=cluster
univ.head()
```
