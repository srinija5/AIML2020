```python
num_col = upvotes.columns[upvotes.dtypes!='category']
```

```python
val = 1
for i in range(len(num_col)):
    for j in range(len(num_col)):
        
        plt.scatter(upvotes[num_col[i]],upvotes[num_col[j]])
        plt.title(num_col[i]+'_'+num_col[j])
        val += 1
```
